## Механические свойства из кривой деформации

Расчёт механических свойств по данным одноосного растяжения:

- **Модуль Юнга (E)** — наклон линейного участка кривой (линейная регрессия)
- **Предел текучести (σ_y)** — напряжение при 0.2% остаточной деформации (метод смещения)
- **Предел прочности (UTS)** — максимальное напряжение на кривой
- **Деформация при UTS** — значение strain при максимальном напряжении
- **Максимальная деформация** — конечная деформация в эксперименте
- **Удельная вязкость разрушения** — площадь под кривой напряжение-деформация

```python
import numpy as np
from scipy import stats

strain = df_ss['Strain'].values
stress = df_ss['Stress_GPa'].values

uts_idx = np.argmax(stress)
uts = stress[uts_idx]
strain_at_uts = strain[uts_idx]
max_strain = strain[-1]
toughness = np.trapezoid(stress, strain)

elastic_end = uts_idx + 1
if elastic_end < 2:
    elastic_end = len(strain)
slope, intercept, r_value, p_value, std_err = stats.linregress(
    strain[:elastic_end], stress[:elastic_end]
)
young_modulus = slope

offset_strain = 0.002
offset_stress = young_modulus * (strain - offset_strain) + intercept
yield_idx = np.argmin(np.abs(stress - offset_stress))
for i in range(len(strain)):
    if stress[i] >= offset_stress[i]:
        yield_idx = i
        break
yield_strength = stress[yield_idx]

print(f"Модуль Юнга (E):          {young_modulus:.2f} ГПа")
print(f"Предел текучести (σ_y):   {yield_strength:.2f} ГПа")
print(f"Предел прочности (UTS):   {uts:.2f} ГПа")
print(f"Деформация при UTS:       {strain_at_uts:.6f}")
print(f"Максимальная деформация:  {max_strain:.6f}")
print(f"Удельная вязкость:        {toughness:.4f} ГПа")
print(f"R² линейной аппроксимации: {r_value**2:.6f}")
```

### Таблица механических свойств

```python
mech_props = pd.DataFrame([
    {'Свойство': 'Модуль Юнга (E)', 'Значение': f'{young_modulus:.2f}', 'Единица': 'ГПа'},
    {'Свойство': 'Предел текучести (σ_y)', 'Значение': f'{yield_strength:.2f}', 'Единица': 'ГПа'},
    {'Свойство': 'Предел прочности (UTS)', 'Значение': f'{uts:.2f}', 'Единица': 'ГПа'},
    {'Свойство': 'Деформация при UTS', 'Значение': f'{strain_at_uts:.6f}', 'Единица': '—'},
    {'Свойство': 'Максимальная деформация', 'Значение': f'{max_strain:.6f}', 'Единица': '—'},
    {'Свойство': 'Удельная вязкость разрушения', 'Значение': f'{toughness:.4f}', 'Единица': 'ГПа'},
    {'Свойство': 'R² линейной аппроксимации', 'Значение': f'{r_value**2:.6f}', 'Единица': '—'},
])

print(mech_props.to_string(index=False))
```

### Кривая деформации с линейной аппроксимацией

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(strain, stress, 'b-o', linewidth=1.5, markersize=4, label='Напряжение')

elastic_strain = np.linspace(strain[0], strain[elastic_end - 1], 100)
elastic_line = young_modulus * elastic_strain + intercept
ax.plot(elastic_strain, elastic_line, 'r--', linewidth=1.5,
        label=f'Линейная аппроксимация (E = {young_modulus:.1f} ГПа)')

ax.axvline(x=strain_at_uts, color='green', linestyle=':', alpha=0.7,
           label=f'UTS = {uts:.2f} ГПа')
ax.axhline(y=uts, color='green', linestyle=':', alpha=0.7)

ax.set_xlabel('Деформация (Strain)', fontsize=12)
ax.set_ylabel('Напряжение (ГПа)', fontsize=12)
ax.set_title('Кривая деформации — одноосное растяжение HEA Al-Co-Cr-Fe-Ni-Ti', fontsize=13)
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
